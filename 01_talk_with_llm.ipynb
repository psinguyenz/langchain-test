```python
# Completion model

from langchain_openai import OpenAI

llmModel = OpenAI()
```

```python
response = llmModel.invoke("Tell me one fun fact about Kennedy family")
```

```python
print("response")
```

```python
# Streaming output
for chunk in llmModel.stream(
    "Tell me one fun fact about the Kennedy family. I need detail response"
):
    print(chunk, end="", flush=True)
```

```python
from langchain_openai import ChatOpenAI

chatModel = ChatOpenAI(model="gpt-4.1-mini")
```

```python
messages = [
    ("system", "You are an expert AI."),
    ("human", "Who are you, what's your model, like gpt 4.1 or 5.1 or something... and current version."),
]
response = chatModel.invoke(messages)
```
